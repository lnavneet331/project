```python
import pandas as pd

data = []

# Iterate through the text files
for i in range(1, 95):
    file_path = f"EmotioNet/emotioNet_challenge_files_server_challenge_1.2_aws/dataFile_1{i:03d}.txt"
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip().split('\t')
            url_stable_copy = line[0]
            url_original = line[1]
            annotations = line[2:]
            
            # Append the data to the list
            data.append({
                'url_stable_copy': url_stable_copy,
                'url_original': url_original,
                'annotations': annotations
            })

# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
# Replace '999' with None for missing or occluded annotations
df = df.replace('999', None)
```

```python
import requests
from PIL import Image
import os

# Directory to store downloaded images
image_dir = 'emotionet_images'
os.makedirs(image_dir, exist_ok=True)

# Download images and save them with appropriate file names
for index, row in df.iterrows():
    image_url = row['url_original']
    image_path = f"{image_dir}/{index}.jpg"
    
    try:
        response = requests.get(image_url, stream=True)
        response.raise_for_status()
        
        with open(image_path, 'wb') as file:
            file.write(response.content)
            
        # Open the image to verify if it's valid
        Image.open(image_path)
        
    except Exception as e:
        # Handle any errors that occur during downloading or opening the image
        print(f"Error downloading image at index {index}: {e}")
        # You can choose to use the stable copy URL instead if available: row['url_stable_copy']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Reset the index of the DataFrames
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
```
